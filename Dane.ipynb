```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/home/bartosz/Pobrane/marketing_dataset/train.csv")
```

```python
sns.factorplot('job', data=df, kind='count')
```

```python
fig, ax = plt.subplots(1,2, figsize = (18, 8))
df["job"].value_counts().plot.bar(color = "black", ax = ax[0])
ax[0].set_ylabel("wynik")
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize = (10, 8))
df["job"].value_counts().plot.bar(color = "yellow", ax = ax[0])
ax[0].set_ylabel("wynik")
plt.show()
```
